```python
import matplotlib.pylab as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
#gambar ke binary
im_gray = cv2.imread("./rmv.jpg", cv2.IMREAD_GRAYSCALE)
(thresh, im_bw) = cv2.threshold(im_gray, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
thresh = 127
im_bw = cv2.threshold(im_gray, thresh, 255, cv2.THRESH_BINARY)[1]
thresh = 127
cv2.imwrite("bw_rmv.jpg", im_bw)

plt.subplot(121),plt.imshow(im_gray, cmap='gray'),plt.title('Original')
#plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(im_bw, cmap='gray'),plt.title('Black White')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#erosion
im = cv2.imread("./bw_rmv.jpg")
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(im,kernel,iterations = 1)
plt.figure(figsize=(20,30))
plt.subplot(121),plt.imshow(im, cmap='gray'),plt.title('Original')
#plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(erosion, cmap='gray'),plt.title('Black White')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#dilation
im = cv2.imread("./bw_rmv.jpg")
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(im,kernel,iterations = 1)
plt.figure(figsize=(20,30))
plt.subplot(121),plt.imshow(im, cmap='gray'),plt.title('Original')
#plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dilation, cmap='gray'),plt.title('Black White')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#opening
im = cv2.imread("./bw_rmv.jpg")
kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(im, cv2.MORPH_OPEN, kernel)
plt.figure(figsize=(20,30))
plt.subplot(121),plt.imshow(im, cmap='gray'),plt.title('Original')
#plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(opening, cmap='gray'),plt.title('Black White')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#closing
im = cv2.imread("./bw_rmv.jpg")
kernel = np.ones((5,5),np.uint8)
closing = cv2.morphologyEx(im, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(20,30))
plt.subplot(121),plt.imshow(im, cmap='gray'),plt.title('Original')
#plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(closing, cmap='gray'),plt.title('Black White')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#tophat
im = cv2.imread("./bw_rmv.jpg")
kernel = np.ones((5,5),np.uint8)
tophat = cv2.morphologyEx(im, cv2.MORPH_TOPHAT, kernel)

plt.figure(figsize=(20,30))
plt.subplot(121),plt.imshow(im, cmap='gray'),plt.title('Original')
#plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(tophat, cmap='gray'),plt.title('Black White')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#blackhat
im = cv2.imread("./bw_rmv.jpg")
kernel = np.ones((5,5),np.uint8)
blackhat = cv2.morphologyEx(im, cv2.MORPH_BLACKHAT, kernel)

plt.figure(figsize=(20,30))
plt.subplot(121),plt.imshow(im, cmap='gray'),plt.title('Original')
#plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blackhat, cmap='gray'),plt.title('Black White')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
from skimage import morphology, filters
from skimage import data
```

```python
#cek gambar ke binary
im = mpimg.imread("./bw_rmv.jpg")
binary = im > filters.threshold_otsu(im)
np.unique(binary)
```

```python
#skeleton
skeleton = morphology.skeletonize(binary)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 12), sharex=True, sharey=True)
ax = axes.ravel()

ax[0].imshow(binary, cmap='gray')
ax[0].set_title("original", fontsize=20)
ax[0].axis("off")

ax[1].imshow(skeleton, cmap='gray')
ax[1].set_title("skeleton", fontsize=20)
ax[1].axis("off")

fig.tight_layout()
plt.show()
```

```python
#thinning
thinned = morphology.thin(binary)
thinned_partial = morphology.thin(binary, max_iter=25)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 12), sharex=True, sharey=True)
ax = axes.ravel()

ax[0].imshow(thinned, cmap='gray')
ax[0].set_title("thinned", fontsize=20)
ax[0].axis("off")

ax[1].imshow(thinned_partial, cmap='gray')
ax[1].set_title("partially thinned", fontsize=20)
ax[1].axis("off")
           
fig.tight_layout()
plt.show()
```
