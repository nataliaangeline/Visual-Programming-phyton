```python
#perbandingan
from scipy import misc, ndimage
from matplotlib.pyplot import imread

im4 = imread ("./flower.jpg")
k = 7 # 7x7 kernel
im_box = ndimage.uniform_filter(im4, size=(k,k,1))
s = 2 # sigma value
t = ((k - 1)/2)-0.5/s #truncate
im_gaussian = ndimage.gaussian_filter(im4, sigma=(s,s,0), truncate=t)
fig = plt.figure(figsize=(30,10))
plt.subplot(131), plt.imshow(im4),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(im_box),plt.title('with the box filter')
plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(im_gaussian),plt.title('with the gaussian')
plt.xticks([]), plt.yticks([])
plt.show()
```
